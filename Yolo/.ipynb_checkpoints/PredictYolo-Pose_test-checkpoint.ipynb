```python
import cv2
import matplotlib.pyplot as plt 
```

```python
from ultralytics import YOLO
model = YOLO('best.pt')
```

```python
video_path = '6.mp4'

cap = cv2.VideoCapture(video_path)

while True:

    success, frame = cap.read()

    if not success:
        break

    results = model(frame, conf=0.5)
    boxes = results[0].boxes.xyxy.tolist()
    test = frame.copy()
    for box in boxes:
        x1, y1, x2, y2 = box
        cv2.rectangle(test, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 10)
    
    
    cv2.imshow("Image", test)
    
    key = cv2.waitKey(10)
    
    if key == 27:
        break
    
cv2.destroyAllWindows()
cap.release()
```
